```python
import pandas as pd
import os
from glob import glob

folder = "../data/Sensex30_Data"

csv_files = glob(os.path.join(folder, "*.csv"))

master_df = pd.DataFrame()

for file in csv_files:
    df = pd.read_csv(file)

    stock_name = os.path.splitext(os.path.basename(file))[0]
    df["Stock"] = stock_name

    master_df = pd.concat([master_df, df], ignore_index=True)

master_df.reset_index(drop=True, inplace=True)

master_df.to_csv("../data/Sensex30_Master.csv", index=False)

print(master_df.head())
```

```python
master_df.rename(columns={"Price": "Date"}, inplace=True)

master_df = master_df[master_df["Date"] != "Ticker"]

numeric_cols = ["Close", "High", "Low", "Open", "Volume"]
master_df[numeric_cols] = master_df[numeric_cols].apply(pd.to_numeric)

master_df = master_df.sort_values(["Stock", "Date"])

master_df.dropna(inplace=True)

```

```python
master_df["Date"] = pd.to_datetime(master_df["Date"])
```

```python
master_df = master_df[["Stock", "Date", "Close", "High", "Low", "Open", "Volume"]]
```
